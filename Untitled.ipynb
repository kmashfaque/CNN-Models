```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import numpy as np

def draw_block(ax, origin, size, label, color):
    x, y, z = origin
    dx, dy, dz = size
    xx = [x, x+dx, x+dx, x, x]
    yy = [y, y, y+dy, y+dy, y]
    kwargs = {'alpha': 0.6, 'color': color}

    ax.plot3D(xx, yy, [z]*5, **kwargs)  # bottom
    ax.plot3D(xx, yy, [z+dz]*5, **kwargs)  # top

    for i in range(4):
        ax.plot3D([xx[i], xx[i]], [yy[i], yy[i]], [z, z+dz], **kwargs)

    ax.text(x+dx/2, y+dy/2, z+dz+0.5, label, ha='center', fontsize=10, color='black')

# Create figure
fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(elev=20, azim=-50)

# Define block locations and sizes
blocks = {
    "Input Feature Map": ((0, 0, 0), (3, 3, 1), "skyblue"),
    "AvgPool (Channel)": ((4, 1, 0), (2, 1, 1), "orange"),
    "MaxPool (Channel)": ((4, -1, 0), (2, 1, 1), "orange"),
    "MLP Shared": ((7, 0, 0), (2, 2, 1), "lightgreen"),
    "Sigmoid Gate (Channel)": ((10, 0, 0), (2, 2, 1), "lightcoral"),
    "Channel Attention Output": ((13, 0, 0), (3, 3, 1), "cyan"),

    "AvgPool (Spatial)": ((16, 0, 0), (1, 1, 1), "orange"),
    "Conv 7x7": ((18, 0, 0), (2, 2, 1), "lightgreen"),
    "Sigmoid Gate (Spatial)": ((21, 0, 0), (2, 2, 1), "lightcoral"),
    "Final Output": ((24, 0, 0), (3, 3, 1), "lime")
}

# Draw blocks
for label, (origin, size, color) in blocks.items():
    draw_block(ax, origin, size, label, color)

# Clean axes
ax.set_xlim(0, 30)
ax.set_ylim(-5, 5)
ax.set_zlim(0, 5)
ax.axis('off')

plt.title("3D CBAM Attention Architecture (Channel + Spatial)", fontsize=14)
plt.tight_layout()
plt.show()
```
